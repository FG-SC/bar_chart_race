# Intro

This notebook follows the tutorial on https://www.dexplo.org/bar_chart_race/tutorial/

```python
import pandas as pd
import bar_chart_race as bcr 

chains_tvl = pd.read_csv('chains.csv', index_col=0)
chains_tvl['Date']= pd.to_datetime(chains_tvl['Date'])
chains_tvl= chains_tvl.set_index('Date', drop=True)
chains_tvl
```

```python
chains_tvl.fillna(0, inplace=True)
```

```python

# Documentation: https://pypi.org/project/bar-chart-race/
bcr.bar_chart_race(
    df=chains_tvl.loc['2024-01-01':],
    filename='btc_sidechains.mp4',
    orientation='h',
    sort='desc',
    n_bars=5,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=4,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Overall TVL: {v.sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    period_length=150, #half second?
    figsize=(12, 6),
    dpi=150,
    cmap='dark12',
    title='BTC sidechains TVL in USD (YTD)',
    bar_label_size=7,
    tick_label_size=7,
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)
```
